```python
# import library
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# load data
# Read the MNIST dataset in keras built into tensorflow
mnist = tf.keras.datasets.mnist.load_data()
train, test = mnist
(x_train, y_train),(x_test, y_test) = mnist

# visualize train data with matplotlib
plt.imshow(x_train[0])
```

```python
# machine learning

# normalize data
x_train = x_train / 255.0
x_test = x_test / 255.0

# set input layer
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Input((28, 28)))  # Input layer is 28 * 28
model.add(tf.keras.layers.Flatten())  # Convert to a one-dimensional array

# set middle layer
model.add(tf.keras.layers.Dense(128))  # Fully combined into 128
model.add(tf.keras.layers.Activation(tf.keras.activations.relu))  # Set the activation function of the middle layer
model.add(tf.keras.layers.Dropout(0.2))  # 20% drop out

model.add(tf.keras.layers.Dense(10)) # Set output layer
model.add(tf.keras.layers.Activation(tf.keras.activations.softmax)) # Set the activation function of the output layer

# Define how to train the model
model.compile(
   optimizer=tf.keras.optimizers.Adam(),
   loss=tf.keras.losses.sparse_categorical_crossentropy,
   metrics=[tf.keras.metrics.sparse_categorical_accuracy]
)
model.fit(x_train, y_train, epochs=5)

# train data 5 times
model.evaluate(x_test, y_test)
```

```python
# visualize test data with matplotlib
test_data = x_test[0]
plt.imshow(test_data)
```

```python
# try to predict
print(test_data.shape)
print(test_data[np.newaxis].shape)

pred = model.predict(test_data[np.newaxis])
np.argmax(pred)
```
